```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
genero = ["Masculino", "Femenino"]
clase_social = ["Pobre", "Clase media", "Rico"]
```

```python
n = 500

genero_data = []
clase_social_data = []

for i in range(0,n):
    genero_data.append(np.random.choice(genero))
    clase_social_data.append(np.random.choice(clase_social))
    
genero_data[1:10]
```

```python
# Hago una distribución normal de las siguientes variables
altura = (170 + 30 * np.random.randn(n))
peso = 70 + 30 * np.random.randn(n)
edad = np.absolute(np.round(50 + 30 * np.random.randn(n)))
salario = np.absolute(15000 + 10000 * np.random.randn(n))
```

```python
data = pd.DataFrame(
    {
        "Genero":genero_data,
        "Clase social":clase_social_data,
        "Altura": altura,
        "Peso": peso,
        "Edad": edad,
        "Salario": salario
    }
)

data.head(10)
```

## Agrupación de Datos

```python
grouped_genero = data.groupby("Genero")
```

```python
grouped_genero.groups
```

```python
for nombre, groups in grouped_genero:
    print(nombre)
    print(groups)
```

## Transformación de variables

```python
zscore = lambda x : (x - x.mean()/x.std())
```

```python
double_group.transform(zscore)
```

```python
z_group = double_group.transform(zscore)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(z_group["Edad"])
```

```python
fill_na_mean = lambda x : x.fillna(x.mean())
```

```python
double_group.transform(fill_na_mean)
```
