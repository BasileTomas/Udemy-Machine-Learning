# Simulación de Monte Carlo
* Genero dos números aleatorios uniformes ('x' e 'y') entre 0 y 1.
* Calculo usando x^2 + y^2:
    * Si obtengo un resultado menor a 1 -> Estoy dentro del circulo.
    * Si obtengo un resultado mayor a 1 -> Estoy fuera del circulo.
* Calculo el número total de veces que está dentro del círculo y lo divido
  por la probabilidad de caer dentro del círculo.
* Usando esa probabilidad trato de aproximar el valor a pi.
* Repito el experimento determinada cantidad de veces.
*

```python
import random
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def calculo_monteCarlo(n, n_exp):
    pi_avg = 0
    pi_value_list = []
    for i in range(n_exp):
        value = 0
        x = np.random.uniform(0,1,n).tolist()
        y = np.random.uniform(0,1,n).tolist()
        for j in range(n):
            z = np.sqrt(x[j] * x[j] + y[j] * y[j]) 
            if z<=1:
                value +=1
        float_value = float(value)
        pi_value = float_value * 4 / n
        pi_value_list.append(pi_value)
        pi_avg += pi_value

    pi = pi_avg/n_exp
    print(pi)
    figura = plt.plot(pi_value_list)
    return(pi, figura)
```

```python
calculo_monteCarlo(100000,200)
```
