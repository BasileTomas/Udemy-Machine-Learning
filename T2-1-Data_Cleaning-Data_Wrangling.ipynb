Selección de datos importantes

```python
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
mainpath='../python-ml-course/datasets/'
filename='customer-churn-model/Customer Churn Model.txt'
fullpath=os.path.join(mainpath, filename)
data = pd.read_csv(fullpath)
```

```python
data.head()
```

## Subconjunto de datos:

```python
#Si quiero ver solamente los datos de alguna columna:
account_length=data["Account Length"]
account_length.head()
```

```python
#Muestro una lista con todas las columnas del Data Frame
all_columns_list=data.columns.values.tolist()
all_columns_list
```

#### Hago el proceso de selección tomando siempre el conjunto de mas chico


```python
desired_columns=['Account Length','VMail Plan','Day Calls']
```

```python
#Digo que sublist incluye todas las columnas menos "desired_columns"
sublist=[x for x in all_columns_list if x not in desired_columns]
sublist
```

#### Imprimo el nuevo data frame con las columnas que queria

```python
subset=data[sublist]
subset.head()
```

#### Selecciono partes del data frame

```python
data[1:30]
```

```python
### Creo un data frame filtrando por estado
data1=data[data["State"] == "LA"]
data1
```

```python
# Tambien puedo filtrar por mas de una condicion usando el AND y OR
data2=data[(data["State"] == "LA") & (data["Eve Calls"] > 100)]
data2
```

```python
# Tambien puedo filtrar por mas de una condicion usando el AND y OR
data3=data[(data["State"] == "LA") | (data["Eve Calls"] > 100)]
data3
```

```python
# En lugar de imprimir el data frame puedo solamente ver la cantidad de datos:
data1.shape
```

```python
data2.shape
```

### Otra manera de filtrar datos sin condicionales es: "iloc" "loc"

```python
# Indistintamente el orden que se lo coloque el resultado es el mismo
subset_first_50=data[:51][["Day Mins", "Night Mins","Account Length"]]
subset_first_50
```

```python
# Tener en cuenta que primero toma las filas luego las columnas
# Ademas "iloc" usa ubicacion pero no etiquetas de columnas para filtrar
data.iloc[1:10, 3:6]
```

```python
data.iloc[[1,5,6,32,200],[2,5,10]]
```

```python
# En cambio el "loc", admite las etiquetas de las columnas
data.loc[[1,5,6,32,200],["Area Code", "VMail Plan","Eve Mins"]]
```

#### De esta manera creo columnas nuevas

```python
data["Total Mins"]=data["Day Mins"] + data["Night Mins"] + data["Eve Mins"]
```
